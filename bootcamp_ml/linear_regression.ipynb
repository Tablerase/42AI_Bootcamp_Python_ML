```python
import numpy as np
import matplotlib.pyplot as plt

data = np.recfromcsv('./ML_01/are_blue_pills_magics.csv')
x = data['micrograms']
y = data['score']

plt.scatter(x, y, marker='o', alpha=0.6, label="data points")
plt.legend()
```

```python
w = -9
b = 90
params = np.array([b, w]) # because X [1 x]
X = x[:, np.newaxis]

y_hat = np.dot(np.hstack((np.ones((x.shape[0], 1)), X)), params)
plt.scatter(x, y, marker='o', alpha=0.6, label="data points")
plt.plot(x, y_hat, color='red', label="prediction line")
```
